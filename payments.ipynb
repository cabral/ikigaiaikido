```python

# Convert the 'Amount' column to numeric, coercing errors to NaN
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

# Define the amounts to filter
amounts_to_filter = [350.00, 700.00, 1050.00]

# Filter the DataFrame for specific amounts
df_filtered = df[df['Amount'].isin(amounts_to_filter)]

# Select only 'Sendername' and 'Amount' columns
df_result = df_filtered[['Sendername', 'Amount']]

# Get the list of Sendernames in df_result
sendernames = df_result['Sendername'].unique()

# Find all transactions for these Sendernames with other amounts
df_other_amounts = df[(df['Sendername'].isin(sendernames)) & (~df['Amount'].isin(amounts_to_filter))]

# Select 'Sendername' and 'Amount' columns
df_other_amounts = df_other_amounts[['Sendername', 'Amount']]

# Reset the index for both DataFrames (optional)
df_result.reset_index(drop=True, inplace=True)
df_other_amounts.reset_index(drop=True, inplace=True)

# Display df_result and df_other_amounts side by side
print("Transactions with Amounts 350.00, 700.00, or 1050.00:")
print(df_result)

print("\nOther Transactions for the same Sendernames with different Amounts:")
print(df_other_amounts)
```
